### Exercise 1: Read the Total profit of all months and show it using a line plot.
Total profit data provided for each month. The enerated line plot must include the following properties:<br>
- X label name = Month Number
- Y label name = Total profit

The line plot graph should look like this:


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Write your code here
url = 'https://pynative.com/wp-content/uploads/2019/01/company_sales_data.csv'

data = pd.read_csv(url)
```

```python
data_plt = plt
data_plt.figure(figsize=(7,5))
data_plt.title('Company profit per Month')
data_plt.plot(data['month_number'], data['total_profit'])
data_plt.xlabel('Month numbere')
data_plt.ylabel('Total profit')
data_plt.xticks(data['month_number'])
data_plt.show()
```

### Exercise 8: Calculate the total sale data for last year for each product and show it using a Pie chart.
Note: In the Pie chart you should display the Number of units sold per year for each product in percentage.

The Pie chart should look like this:


```python
# Write your code here
#pie chart distribution of sales data(number of units sold).
#chart labels = product names = data.[all product names]
#chart distribution = Sum of units sold in each month(sum for all products)

products = ['facecream', 'facewash', 'toothpaste', 'bathingsoap', 'shampoo', 'moisturizer']
total_unit_list = [data[product].sum() for product in products]
# for product in products:
#     total_unit_list.append(data[product].sum())

plt.figure(figsize=(12,10))
plt.axis("equal")
plt.pie(total_unit_list, labels=products, autopct='%1.1f%%', textprops={'color':"black",'fontsize':15})
plt.legend(loc='lower right')
plt.title('Sales data')
plt.show()
```
