```python
# Imports
import matplotlib
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import datasets
from sklearn import preprocessing
from sklearn.model_selection import train_test_split, cross_validate,GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.preprocessing import LabelEncoder
import sklearn.tree as tree
import numpy as np
import pandas as pd
# import pydotplus
%matplotlib inline
```

```python
data = datasets.load_iris()
df = pd.DataFrame(data = data.data)
df.columns = data.feature_names
df['class'] = data.target
df['target'] = data.target
df['class'] = df['class'].apply(lambda x: data['target_names'][x])
df.sample(n=10, random_state=1)
```

```python
# Assigning X and y
X = df[['sepal length (cm)','sepal width (cm)','petal length (cm)', 'petal width (cm)']]
y = df['target']
```

```python
# Train Test Split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size = 0.20, random_state=0)
```

```python
rfc = RandomForestClassifier(random_state=4)
```

```python
rfc.fit(x_train, y_train)
```

```python
rfc.score(x_test,y_test)
```

```python
df1 = pd.read_csv("zoo.csv", sep=',')
df1.info()
```

```python
## Since Animal names are unique we can drop them or we can perform label encoding.
X1 = df1.drop(['class_type'], axis=1)
y1 = df1['class_type']
```

```python
labelencoder = LabelEncoder()
X1_labled = X1.copy()

X1_labled['animal_name'] = labelencoder.fit_transform(X1['animal_name'])
```

```python
# Train Test Split
x1_train,x1_test,y1_train,y1_test = train_test_split(X1_labled,y1,test_size = 0.20, random_state=3)
x1_train
```

```python
rfc1 = RandomForestClassifier(criterion = 'entropy', max_depth =5, min_samples_leaf= 1, min_samples_split= 2, n_estimators=3, bootstrap=True)
```

```python
rfc1.fit(x1_train, y1_train)
```

```python
rfc1.score(x1_test,y1_test)
```
